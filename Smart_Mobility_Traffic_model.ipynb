```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('/content/smart_mobility_dataset.csv')
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
print(f'data overall{x}')

print(f'\ntarget variable{y}')
```

```python
x.info()
```

```python
y

```

```python
obj=x.select_dtypes(include='object').columns
print(obj)
```

```python
from sklearn.preprocessing import LabelEncoder


#Create a LabelEncoder instance
label_encoder=LabelEncoder()

#List of catagorical columns to encode
categorical_columns=obj

#loop through each categorical column and apply labelencoding
for col in categorical_columns:
  x[col]=label_encoder.fit_transform(x[col])

# Display the updated DataFrame
print("\nUpdated Features (x) after Label Encoding:")
print(x)
```

```python
y=label_encoder.fit_transform(y)
print(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
LogisticRegression_model=LogisticRegression(random_state=0)
LogisticRegression_model.fit(x_train_scaled,y_train)
```

```python
weather_prediction=LogisticRegression_model.predict(x_test_scaled)
print("LogisticRegresstion:")
print("Accuracy:",accuracy_score(y_test,weather_prediction))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
DecisionTreeClassifier_model=DecisionTreeClassifier(random_state=0)
DecisionTreeClassifier_model.fit(x_train_scaled,y_train)

```

```python
weather_prediction=DecisionTreeClassifier_model.predict(x_test_scaled)
print("DecisionTreeClassifier:")
print("Accuracy:",accuracy_score(y_test,weather_prediction))
```

```python
RandomForestClassifier_model=RandomForestClassifier(random_state=0)
RandomForestClassifier_model.fit(x_train_scaled,y_train)
```

```python
weather_predict=RandomForestClassifier_model.predict(x_test_scaled)
print("RandomForestClassifier:")
print("Accuracy:",accuracy_score(y_test,weather_predict))
```
